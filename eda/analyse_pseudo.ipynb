```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../testdata.csv')
# df = pd.read_csv('../macaulaylibrary.csv')
scores = np.zeros(len(df))
for idx, row in df.iterrows():
    scores[idx] = np.array(row[2].replace('[','').replace(']','').replace(' ','').split(','), dtype='float').max()

scores.sort()
plt.plot(scores)
plt.ylim([0.999, 1])
```
